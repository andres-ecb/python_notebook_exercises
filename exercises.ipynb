## 1. Práctica Uno: Programas Secuenciales
Desarrollar un programa que permita registrar las notas de una asignatura X, de uno o varios estudiantes.

El programa debe solicitar, inicialmente, la cantidad de estudiantes a registrar y, para cada uno de ellos, pedir el nombre y apellido del estudiante, la cantidad de notas a ingresar (con un mínimo de tres notas) y las notas correspondientes, las cuales deben estar en un rango entre 1.0 y 5.0. En caso de que se ingresen más de cinco notas para un estudiante, el programa deberá considerar únicamente las cinco notas más altas para los cálculos posteriores.

A partir de las notas admitidas, una vez sean validadas, el programa debe calcular: la nota más alta, la más baja, y el promedio de las notas consideradas.

Además, el programa debe calcular una nota definitiva, la cual se obtiene sumando al promedio una bonificación por cantidad de notas ingresadas. Este bono se asigna de la siguiente manera: 
- Si el estudiante registra únicamente el número mínimo de notas requeridas, no recibe bonificación
- Por cada nota adicional al mínimo (tres), hasta un máximo de cinco notas consideradas, se suma un bono fijo de 0.05 al promedio
- Si el estudiante registra más de cinco notas, el bono máximo aplicable será el correspondiente a cinco notas.

La nota mínima para aprobar es 3. Con base en el promedio, el programa debe indicar si el estudiante aprueba (promedio mayor o igual a 3) o reprueba (promedio menor que 3).

Finalmente, toda la información generada debe almacenarse en un dataframe de Pandas denominado ***notas_definitivas*** y mostrarse en pantalla.

```python
import pandas as pd

#Definición de las cantidades máxima y mínima de notas por estudiante y de la bonificación por nota adicional
MIN_NOTAS = 3
MAX_NOTAS = 5
BONO_POR_NOTA = 0.05

nombres = []
nota_maxima = []
nota_minima = []
promedios = []
nota_definitiva = []
bonificacion = []
estado = []
cantidades_nota = [] #Las listas que contendrán los datos; una por cada columna del dataframe a crear

print("☰ GESTOR DE NOTAS ESTUDIANTILES ☰")
print(f"Debe ingresar, al menos, {MIN_NOTAS} notas por estudiante")

while True:
    try:
        cant_estudiantes = int(input("Cantidad de estudiantes a registrar: "))
        if cant_estudiantes > 0:
            break
        else:
            print("Debe ingresar al menos un estudiante") # Comprueba que la cantidad de est. no quede en cero
    except ValueError:
        print("Ingrese un número válido") # Comprueba que solo se ingresen números

for e in range(cant_estudiantes): # Ciclo for para registrar las notas para cada estudiante
    
    while True:
        nombre = input("Nombre del estudiante: ").strip()
        if nombre == "":
            print("El nombre no puede estar en blanco")
        elif any(char.isdigit() for char in nombre): # Valida que el nombre no contenga números
            print("El nombre no puede contener números")
        else:
            nombres.append(nombre)
            break

    while True:
        try:
            cant_notas = int(input(f"Cantidad de notas para {nombre}: "))
            if cant_notas >= MIN_NOTAS: # Comprueba que la cantidad de notas sea igual o mayor que tres
                cantidades_nota.append(cant_notas)
                break
            else:
                print(f"Recuerde ingresar al menos {MIN_NOTAS} notas")
        except ValueError:
            print("Ingrese un número válido") # Comprueba que solo se ingresen números
    print(f"Estudiante No. {e + 1}: ", nombre, " - ", cant_notas, "notas")

    notas = [] # Lista para aglutinarlas notas por cada estudiante
    n = 0
    
    while n < cant_notas:
        try:
            nota = float(input(f"Ingrese la nota No. {n + 1}: "))
            if 1.0 <= nota <= 5.0: # Comprueba que las notas ingresadas estén dentro del rango permitido
                notas.append(nota)
                n += 1
            else:
                print("La nota debe estar entre 1.0 y 5.0.")
        except ValueError:
            print("Ingrese un valor numérico válido.")
    
    if len(notas) > MAX_NOTAS: # Tiene en cuenta solo las cinco notas más altas, si hay más de cinco
        notas.sort(reverse=True) # Como sort() organiza de forma ascendente, debe reversarse para tomar las notas cinco más altas
        notas = notas[:MAX_NOTAS]

    promedio = round(sum(notas) / len(notas), 2)

    notas_efectivas = min(cant_notas, MAX_NOTAS)
    bono = (notas_efectivas - MIN_NOTAS) * BONO_POR_NOTA
    bono = max(0, bono)
    nota_final = round(promedio + bono, 2) #La nota definitiva es una métrica derivada del promedio y la bonificación

    nota_maxima.append(max(notas))# Los datos entran a las listas definidas al inicio
    nota_minima.append(min(notas))
    promedios.append(promedio) 
    bonificacion.append(bono)
    nota_definitiva.append(nota_final)
    estado.append("Aprobado" if nota_final >= 3 else "Reprobado") # Se asigna el estado dependiendo de la nota def.

#Aquí termina el ciclo for

notas_definitivas = pd.DataFrame({ # Se crea el dataFrame, mediante un diccionario, con las listas como columnas, 
    "Nombre": nombres,
    "Nota más Alta": nota_maxima,
    "Nota más Baja": nota_minima,
    "Cantidad de Notas" : cantidades_nota,
    "Promedio": promedios,
    "Bonificación": bonificacion,
    "Nota Definitiva": nota_definitiva,
    "Estado": estado
})

print("\nDataframe 'notas_definitivas':")
notas_definitivas
```

## **3.** Práctica Tres: Programación Orientada a Objetos

Desarrollar un programa que permita ingresar las medidas de los lados de varios triángulos. El programa debe validar que las medidas formen un triángulo correcto, es decir, que la suma de dos lados debe ser mayor que el tercero.

Utilizando una clase ***Triangulo***, determinar si cada triángulo es equilátero, isósceles o escaleno, según la medida de sus lados.

El programa debe solicitar cuántos triángulos se desean ingresar, mostrar la clasificación de cada uno y, al finalizar, presentar un conteo total de triángulos por tipo.

```python
class Triangulo: # Se crea una clase para gestionar el registro del triángulo
    def __init__(self, lado1, lado2, lado3):
        self.lado1 = lado1
        self.lado2 = lado2
        self.lado3 = lado3

    def validez(self): #Método de la clase para validar que el triángulo sea matemáticamente correcto
        return (
            self.lado1 > 0 and self.lado2 > 0 and self.lado3 > 0 and
            self.lado1 + self.lado2 > self.lado3 and
            self.lado1 + self.lado3 > self.lado2 and
            self.lado2 + self.lado3 > self.lado1
        )

    def tipo(self): #Método de la clase para identificar el tipo de triángulo según las medidas de sus lados
        #Por medio de un set{}, y de len(), se determina cuántos lados están repetidos
        tres_lados = len({self.lado1, self.lado2, self.lado3})
        if tres_lados == 1:
            return "Equilátero"
        elif tres_lados == 2:
            return "Isósceles"
        else:
            return "Escaleno"

#Función para gestionar que las medidas de los lados sean valores aceptables
def ingreso_lado(mensaje):
    while True:
        try:
            valor = float(input(mensaje))
            if valor > 0:
                return valor
        except ValueError:
            print("Debe ingresar un número válido")


print("☰ CLASIFICADOR Y CONTADOR DE TRÍANGULOS ◭ ☰")

while True:
    try:
        cant_tg = int(input("¿Cuántos triángulos ingresará?: "))
        if cant_tg > 0:
            break
        print("Debe ingresar al menos un triángulo")
    except ValueError:
        print("Ingrese un número válido")

contador = {
    "Equilátero": 0,
    "Isósceles": 0,
    "Escaleno": 0
}

for i in range(cant_tg):
    while True:
        l1 = ingreso_lado("Medida del 1er lado: ")
        l2 = ingreso_lado("Medida del 2do lado: ")
        l3 = ingreso_lado("Medida del 3er lado: ")
        triangulo = Triangulo(l1, l2, l3)

        if triangulo.validez():
            break
        else:
            print("Triángulo no válido")

    tipo = triangulo.tipo()
    print(f"Triángulo No. {i + 1} / Lados: {l1} - {l2} - {l3} / Es un triángulo", tipo)
    contador[tipo] += 1


print("\nDistribución de Triángulos Ingresados")

print("Triángulos Equiláteros:", contador["Equilátero"])
print("Triángulos Isósceles:", contador["Isósceles"])
print("Triángulos Escalenos:", contador["Escaleno"])
```
